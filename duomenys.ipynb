```python
import pandas as pd
import random
import os
import re
from google.colab import drive
```

```python
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/MyDrive/Rap-bot data/lyrics_dataset.csv")

if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
lyrics = []

artists = df['artist'].unique()
random_artists = random.sample(list(artists), 100)

for artist in random_artists:
    artist_df = df[df['artist'] == artist]
    songs = artist_df['song'].unique()
    random_songs = random.sample(list(songs), min(5, len(songs)))

    for song in random_songs:
        lines = artist_df[artist_df['song'] == song]["lyric"].dropna().tolist()
        lines = [l.strip() for l in lines if str(l).strip() != ""]
        lyrics.extend(lines)
        lyrics.append("<EOS>")
```

```python
text_All = "\n".join(lyrics).lower()

text_All = re.sub(r"[^a-z0-9\s.,!?\'\"\[\]\(\)\-<>]", "", text_All)
text_All = re.sub(r" +", " ", text_All)

print("Bendras simbolių kiekis:", len(text_All))
```

```python
chars = sorted(list(set(text_All)))
char_indices = {c: i for i, c in enumerate(chars)}
indices_char = {i: c for i, c in enumerate(chars)}

print("Unikalių simbolių:", len(chars))
```

```python
with open("/content/drive/MyDrive/rapbot_text_all.txt", "w", encoding="utf-8") as f:
    f.write(text_All)

import json

meta = {
    "chars": chars,
    "char_indices": char_indices,
    "indices_char": indices_char
}

with open("/content/drive/MyDrive/rapbot_char_meta.json", "w", encoding="utf-8") as f:
    json.dump(meta, f)
```
